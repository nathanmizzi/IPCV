```python
import cv2
import numpy as np

def nothing(x):
    pass

# Creating a black image
image = np.zeros((300,500,3), np.uint8)
# Creating a window
cv2.namedWindow('ColorPicker')

# Creating trackbars
cv2.createTrackbar('R', 'ColorPicker',0,255,nothing)
cv2.createTrackbar('G', 'ColorPicker',0,255,nothing)
cv2.createTrackbar('B', 'ColorPicker',0,255,nothing)
cv2.createTrackbar('A', 'ColorPicker',0,1,nothing)

while(1):
    cv2.imshow('ColorPicker', image)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    r = cv2.getTrackbarPos('R', 'ColorPicker')
    g = cv2.getTrackbarPos('G', 'ColorPicker')
    b = cv2.getTrackbarPos('B', 'ColorPicker')
    a = cv2.getTrackbarPos('A', 'ColorPicker')

    if a == 0:
        image[:] = [0,0,0]

    else:
        image[:] = [b,g,r]

cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.cm as cm

image = cv2.imread('images/lena.jpg')

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(image_rgb)
plt.show()

def decrease_brightness(img,val):
    # Changing the color space to HSV
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Splitting each color channel
    h,s,v = cv2.split(hsv)

    # Increase the brightness value

    # Getting the maximum value that value can be without exceeding 255
    lim = 0 + val
    # if v is going to exceed 255, we set it to 255
    v[v < lim] = 0
        #else
    v[v >= lim] -= val

    #Rejoin (Merge) the channels to hsv
    merged = cv2.merge((h,s,v))
    return cv2.cvtColor(merged, cv2.COLOR_HSV2RGB)

def increase_brightness(img,val):
    # Changing the color space to HSV
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Splitting each color channel
    h,s,v = cv2.split(hsv)

    # Increase the brightness value

    # Getting the maximum value that value can be without exceeding 255
    lim = 255 + val
    # if v is going to exceed 255, we set it to 255
    v[v > lim] = 255
        #else
    v[v >= lim] -= val

    #Rejoin (Merge) the channels to hsv
    merged = cv2.merge((h,s,v))
    return cv2.cvtColor(merged, cv2.COLOR_HSV2RGB)

image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)


plt.axis('off')
plt.imshow(decrease_brightness(image, 50))
plt.show()
```
