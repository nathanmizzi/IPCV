1.	Colour splitting
Write a program which reads/loads an RGB image and displays the red, green, and blue channels separately, each one of them as an RGB image.


```python
# Reading an image
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2

goldfish = cv2.imread("../images/goldfish.jpg")

# Convert from BGR to RGB
goldfishNew = cv2.cvtColor(goldfish, cv2.COLOR_BGR2RGB)

inverted = 255 - goldfishNew

print("Goldfish")
plt.imshow(goldfishNew)
plt.show()

print("Goldfish Red")
ML_R = goldfishNew.copy()

ML_R[:,:,(1,2)] = 0

plt.imshow(ML_R)
plt.show()
print("Goldfish Green")
ML_G = goldfishNew.copy()

ML_G[:,:,(0,2)] = 0

plt.imshow(ML_G)
plt.show()

print("Goldfish Blue")
ML_B = goldfishNew.copy()

ML_B[:,:,(0,1)] = 0

plt.imshow(ML_B)
plt.show()
```

## Question 2

2.	Image Brightening
Write a program which loads an RGB image and brightens each pixel by a value of 5 (current brightness + 5). Do this by applying an addition operation between a matrix and a scalar.


```python
# Reading an image
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2

goldfish = cv2.imread("../images/goldfish.jpg")

# Convert from BGR to RGB
goldfishNew = cv2.cvtColor(goldfish, cv2.COLOR_BGR2RGB)

print("Goldfish")

goldfishNew[:,:] += 5

plt.imshow(goldfishNew)
plt.show()
```

## Question 3

3.	Image trimming
Write a program which loads an RGB image and trims out a margin of 20 pixels from all around the image. Do this by using the colon notation. Display the new image.


```python
# Reading an image
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2

goldfish = cv2.imread("../images/goldfish.jpg")

# Convert from BGR to RGB
goldfishNew = cv2.cvtColor(goldfish, cv2.COLOR_BGR2RGB)

print("Original")
plt.imshow(goldfishNew)
plt.show()

goldfishCropped = goldfishNew[20:-20,20:-20,:]

print("New")
plt.imshow(goldfishCropped)
plt.show()
```

Resizing an Image

```python
# Reading an image
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

lena = cv2.cvtColor(cv2.imread("../images/lena.jpg"), cv2.COLOR_BGR2RGB)
lena = cv2.resize(lena,(200,100))

print("Original")
plt.imshow(lena)
plt.show()
```

```python
# Downscaling an image while keeping the aspect ratio
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

lena = cv2.cvtColor(cv2.imread("../images/lena.jpg"), cv2.COLOR_BGR2RGB)

lena2 = cv2.resize(lena, (round(lena.shape[1]/2), round(lena.shape[0]/2)))

print("Original")
plt.imshow(lena)
plt.show()
Image.fromarray(lena2.astype(np.uint8)).save("../images/original.png")

print("New")
plt.imshow(lena2)
plt.show()
Image.fromarray(lena2.astype(np.uint8)).save("../images/small.png")
```
