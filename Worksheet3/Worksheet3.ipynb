1.	Create a Python program which loads an image, convert it to grayscale and store it an a numpy array.


```python
# Downscaling an image while keeping the aspect ratio
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2

lena = cv2.cvtColor(cv2.imread("../images/lena.jpg"), cv2.COLOR_BGR2GRAY)

lena_np_array = np.array(lena)

print(lena_np_array)

print("\nOriginal")
plt.imshow(lena, cmap="gray", vmin=0, vmax=255)
plt.show()
```
